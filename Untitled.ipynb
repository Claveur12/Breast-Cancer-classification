```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_data = load_breast_cancer()
cancer_data
```

```python
df_cancer = pd.DataFrame(np.c_[cancer_data['data'], cancer_data['target']], columns=np.append(cancer_data['feature_names'], ['targer']))
```

```python
df_cancer
```

```python
df_cancer.head()
```

```python
df_cancer.tail()
```

```python
sns.pairplot(df_cancer, hue = "targer", vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'])
plt.show()
```

```python
sns.countplot(df_cancer['targer'])
plt.show()
```

```python
sns.scatterplot(x= 'mean area', y = 'mean smoothness', hue ='targer', data = df_cancer)
plt.show()
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(df_cancer.corr(), annot = True)
plt.show()
```

```python
X = df_cancer.drop(['targer'], axis =1)
X
```

```python
y = df_cancer['targer']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

```python
y_predict = svc_model.predict(X_test)
```

```python
y_predict
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot=True)
plt.show()
```

```python
min_train = X_train.min()
```

```python
range_train = (X_train - min_train).max()
```

```python
X_train_scaled = (X_train - min_train)/range_train
```

```python
sns.scatterplot(x= 'mean area', y = X_train['mean smoothness'], hue ='targer', data = df_cancer)
plt.show()
```

```python
sns.scatterplot(x= 'mean area', y = X_train_scaled['mean smoothness'], hue ='targer', data = df_cancer)
plt.show()
```

```python
min_test = X_test.min()
range_test = (X_test - min_test).max()
X_test_scaled = (X_test - min_test)/range_test
```

```python
svc_model.fit(X_train_scaled, y_train)
```

```python
y_predict = svc_model.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot=True)
plt.show()
```

```python
print(classification_report(y_test, y_predict))
```

```python
param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel':['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=4)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
grid.best_params_
```

```python
grid.best_index_
```

```python
grid.best_score_
```

```python
grid_predictions = grid.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, grid_predictions)
```

```python
sns.heatmap(cm, annot=True)
plt.show()
```

```python
print(classification_report(y_test, grid_predictions))
```
